**1. Dataset Overview**  
This dataset comprises of 3,100 curated CCRs across 15 disease groups, including over 750 reports on rare diseases.  
  
Each CCR contains 35 metadata entries, with 10 entries for identification, 21 for medical content, and 4 miscellaneous entries, such as reference count.  
  
From these, features such as Signs and Symptoms, Life Style, Family History, Social History, Medical/Surgical History, Comorbidities, Laboratory Values, Pathology, Age, and Gender were aggregated for predicting diagnoses and treatments.   
  
Possible predictive attributes includes diagnosis, interventional therapy and pharmacological therapy prediction.  

The dataset can be accessed at https://figshare.com/collections/MACCRs/4220324/4  
and  
the original paper can be accessed at https://www.nature.com/articles/sdata2018258

**5. Putting It All Together.**  
Below is a complete example to get you started. Feel free to customize and improve upon it! This is just a foundational structure to inspire your own creative and tailored implementation. Please include at least 3 LLMs in your implementation.

```python
import json
import random
import pandas as pd
import argparse
from tqdm import tqdm
import os
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import csv
from google.colab import drive

# Mount your Google Drive
drive.mount('/content/drive', force_remount=True)

def load_dataset_json(filepath):
    with open(filepath, 'r') as file:
        return json.load(file)

def format_case(case):
    attrs_cols = ['Life Style', 'Family History',
       'Social History', 'Medical/Surgical History', 'Signs and Symptoms',
       'Comorbidities',
       'Laboratory Values',
       'Pathology','Age', 'Gender']

    descriptions = "; ".join(f"{col}: {case.get(col, 'NA') if case.get(col) is not None else 'NA'}" for col in attrs_cols)
    diagnosis = case.get('Diagnosis', 'NA') if case.get('Diagnosis') is not None else 'NA'
    cr_number = case.get('CR Number', 'NA') if case.get('CR Number') is not None else 'NA'
    interventional = case.get('Interventional Therapy', 'NA') if case.get('Interventional Therapy') is not None else 'NA'
    pharmacological = case.get('Pharmacological Therapy', 'NA') if case.get('Pharmacological Therapy') is not None else 'NA'

    return cr_number, descriptions, diagnosis, interventional, pharmacological

def pred_zeroshot(case_description, model, tokenizer, device):
    prompt = f"""You are a highly knowledgeable medical expert with expertise in diagnosing a \
    variety of medical conditions. You are presented with a patient's clinical details and are asked to provide \
    a potential diagnosis.
    For a clinical case like this: {case_description}\n, what is the suggested potential diagnosis? \
    Please provide the potential diagnosis, without any style or formatting."""

    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    input_ids = inputs['input_ids']

    input_token_length = input_ids.shape[1]
    output_ids = model.generate(input_ids, max_new_tokens=128, do_sample=False)
    output = tokenizer.decode(output_ids[0][input_token_length:])
    diagnosis = output.replace('\n', '').strip()

    del inputs
    del output

    return diagnosis

# def initialize_results_csv(dataset, results_fp):
#     if os.path.exists(results_fp):
#         results_df = pd.read_csv(results_fp)
#     else:
#         results_df = pd.DataFrame({
#             'cr_number': dataset['CR Number'],
#             'diagnosis': dataset['Diagnosis'],
#             'pharmacological_therapy': dataset['Pharmacological Therapy'],
#             'interventional_therapy': dataset['Interventional Therapy']
#         })
#         results_df.to_csv(results_fp, index=False)

#     return results_df

def main(args_model, y_attr):
    dataset = load_dataset_json('/content/drive/My Drive/Colab Notebooks/Bioinfo201 Winter Project/MACCRs_cardio.json')

    results_fp = '/content/drive/My Drive/Colab Notebooks/Bioinfo201 Winter Project/pred_results.csv'

    tokenizer = AutoTokenizer.from_pretrained(args_model, device_map="auto", return_tensors="pt", use_fast=False)
    model = AutoModelForCausalLM.from_pretrained(args_model, device_map="auto", torch_dtype=torch.float16)
    device = model.device

    with open(results_fp, 'a', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)

        if csvfile.tell() == 0:
            csvwriter.writerow(['cr_number', 'diagnosis', 'pharmacological_therapy',
                                'interventional_therapy', f"pred_{y_attr}_{args_model}"])  # Adjust columns as needed

        for i, case in tqdm(enumerate(dataset)):
            cr_number, descriptions, diag_gt, interventional, pharmacological = format_case(case)
            pred = pred_zeroshot(descriptions, model, tokenizer, device)
            csvwriter.writerow([cr_number, diag_gt, pharmacological, interventional, pred])

if __name__ == "__main__":
    main('bigscience/bloom-1b7', "diagnosis")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_embedding_bert(text, bert_model):
    return bert_model.encode([text])

def cal_cosine_similarity(text1, text2, client, model='small'):
    cos_similarity = cosine_similarity(get_embedding_bert(text1, client), \
                                           get_embedding_bert(text2, client))[0][0]
    return cos_similarity
```
